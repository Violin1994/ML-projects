## Предобработка данных

```python
#заменяем высту потолков медианным значением
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
# расстояние до центра города
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(
    data.groupby('locality_name')['cityCenters_nearest'].transform('median'))
#data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(0)
#data['cityCenters_nearest'] = data['cityCenters_nearest'].round()
#data['cityCenters_nearest'] = data['cityCenters_nearest'].astype(int)
#площадь кухни
data['kitchen_area'] = data['kitchen_area'].fillna(
    data.groupby('total_area')['kitchen_area'].transform('median'))
#жилая площадь
data['living_area'] = data['living_area'].fillna(
    data.groupby('rooms')['living_area'].transform('median'))
#Общая площадь
data['total_area'] = data['total_area'].fillna(
    data.groupby('locality_name')['total_area'].transform('median'))
data['total_area'] = data['total_area'].round()
data['total_area'] = data['total_area'].astype(int)
data
```

```python
#балконы,число парков, число водоемов,студии и апартаменты
data[['balcony', 'parks_around3000', 'studio', 'ponds_around3000', 'is_apartment']] = data[['balcony', 'parks_around3000','studio', 'ponds_around3000', 'is_apartment']].fillna(0)
#замена пропущенных значений в количестве дней размещения объявления
data['days_exposition'] = data['days_exposition'].fillna(0)
```

```python
# меняем тип данных
data['is_apartment'] = data['is_apartment'].astype(int)
data['studio'] = data['studio'].astype(int)
data['balcony'] = data['balcony'].astype(int)
data['parks_around3000'] = data['parks_around3000'].astype(int)
data['ponds_around3000'] = data['ponds_around3000'].astype(int)
data['days_exposition'] = data['days_exposition'].round()
data['days_exposition'] = data['days_exposition'].astype(int)
data['last_price'] = data['last_price'].round()
data['last_price'] = data['last_price'].astype(int)
data
```

## Расчёты и добавление результатов в таблицу

```python
#цена кв метра
data['price_of_the_meter'] = data['last_price'] / data['total_area']
data['price_of_the_meter'] = data['price_of_the_meter'].round()
data['price_of_the_meter'] = data['price_of_the_meter'].astype(int)
#преобразуем время и дату публикации
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['days'] = pd.DatetimeIndex(data['first_day_exposition']).day
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year

#этаж квартиры
def create_floor_category(row):
    if row[8] ==1:
        return 'Первый'
    elif row[8] == row[6]:
        return 'Последний'
    return 'Другой'
data['floor_category'] = data.apply(create_floor_category, axis=1)
# соотношение жилой и общей площади
data['living_ratio'] = data['living_area'] / data['total_area']
#отношение площади кухни к общей
data['kitchen_ratio'] = data['kitchen_area'] / data['total_area']
#проверка данных
data
```

```python
# Перевожу данные по этажу в формат "первый = 1", "другой = 2" и "последний = 3"
def create_floor_category_number(row):
    if row[8] ==1:
        return 1
    elif row[8] == row[6]:
        return 3
    return 2
data['floor_category_number'] = data.apply(create_floor_category_number, axis=1)
data
```
